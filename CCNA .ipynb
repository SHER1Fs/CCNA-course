# **OSI Model & TCP/IP Suite**


 * **Protocol :**
    
     - a set of rules(logical) difining how network devices and software should work .


 - **1- OSI model :**

     - consists of 7 layers : 

         * **Layer 7 (Application) :**

             - closest layer to the end user 
             - interacts with software apps such as (web browesers )
             - identifying communication parteners and synchronizing communication 
            
         * **Layer 6 (Presentation) :**

             - translate between application and network format (translate data to a propraite format )
             - encrypte and decrypte the data 

         * **Layer 5 (session) :**

             - controls sessions between communicating hosts ( local app and remote app )


         * **Layer 4 (Transport) :**

             - Segments and reassembles data for communication between end hosts 
             - provide end-to-end communication 

         * **Layer 3 (Network) :**

             - Segments become Packets 
             - provides connectivity between end hosts on different networrks 
             - provides IP addresses ( logical Address )
             - provides path selection between source and destination
             - Routers operate in this Layer 

         * **Layer 2 (Data link) :**

             - Packets become Frames 
             - provides node-to-node connectivity and data transfer 
             - provides MAC addresses 
             - Switches operate in this layer   

         * **Layer 1 (Physical) :**
          
             - defines physical characteristics of the medium used to transfer data between devices 
             - digital bits(frames) are converted into electrical signals 



 - **2- TCP\IP Suite :**

     * conceptual model and set of communication protocols used in the internet and other networks 

     * similar to OSI model, but with fewer layers (only 4 Layers )  

     * ( Application + Presentation + Session ) Layers ==> ( Application ) Layer 
     
     * Transport Layer is the same as in OSI 

     * Network Layer ==> Internet Layer 

     * ( Data Link + Physical ) Layers ==> Link Layer 

     * this model is commonly used nowadays

# **IPv4 Addressing**

 * IP address are 4 bytes (32 bits) in length , every 8 bits is an octet 
 
 * written in dotted decimal (192.168.1.0/24) , /24 means that the first 24 bits of this IP address represent the network portion of the address  , and the rest of bits (8) represent the host portion 


 * **IPv4 Address Classes :**
  
     - **1- Class A :**

         * first octet (0xxxxxxx) , numeric range (0:127)

         * uses /8 prefix

         * fewer potential networks , more hosts in each
          network 

         * Netmask (255.0.0.0) 


     * **2- Class B :**

         * first octet (10xxxxxx) , numeric range (128.191)

         * uses /16 prefix

         * Netmask (255.255.0.0)

     - **3- Class C :**

         * first octet (110xxxxx) , numeric range (192:223)

         * uses /24 prefix

         * more potential networks , fewer hosts in each
         network

         * Netmask (255.255.255.0)

     * **4- Class D :**

         * first octet (1110xxxx) , numeric range (224:239)
         
         * reserved for multicast addresses 

     - **5- Class E :**

         * first octet (1111xxxx) , numeric range (240:255)
         
         * reserved for experimental uses 



 * **Loopback Addresses :**

     - **Address Range (127.0.0.0 - 127.255.255.255) :**
        
         - used to test the network stack ( think OSI ,TCP/IP model ) on the local device 


 * if the host portion of an IP address is all 0’s , it means it is the **Network address** or network ID 

 * the Network address can not be assigned to a host   

 * if the host portion of an IP address is all 1’s , it means it is the **Broadcast address**     



 * Maximum hosts per network = (2^n) - 2 , n is number of host bits 



 * **EX ==>** IP address of 43.109.23.12/8  , find the following 

     - 1- Network address :

         * 43.0.0.0

     * 2- max no. of hosts in the network :

         - (2^24) - 2   = 16777214

     - 3- network broadcast address :

         * 43.255.255.255

     * 4- first usable address of the network :

         * 43.0.0.1

     * 5- last usable address of the network :

         * 43.255.255.254              

 


 
       
                







# **Switch Interfaces**


 * **Half Duplex :**

     - the device can not send and recieve data at the same time .

     - Rarely used nowadays 

     - used in Hubs in past 

     - **CSMA/CD (Carrier Sense Multiple Access with Collsion Detection ) :**

         - describes how devices avoid collisions in a half-duplex situation and how they react if collisions do occur  

         - before sending frames , devices listen to the collision domain until they detect that other devices are not sending frames 

         - if a collision occurs , the device sends a jamming signal to inform the other device that a collision happened



 * **Full Duplex :**

     - the device can send and recieve data at the same time   


 - **Speed/Duplex Autonegotiation :**

     - interfaces that can run at different speeds have default settings of **speed auto** and **duplex auto**

     - interfaces advertise thier capabilities to the nearby device , and they negotiate the best speed and duplex setting they are both capable of 


 * **Interface Errors :**

     - **2- Giants :**

         * frames that failed CRC check 

     - **4- Frame :**

         * frames that have an incorrect format due to an error 

     - **5- input frames :**

         * total various counters , such as the above 

     - **6- output frames :**

         * frames the switch tried to sent , but failed due to an error 
         
                              




  







# **IPv4 Header**


 * Protocol Data Units ( PDUs ) :

     - Data 
     - Segement 
     - Packet 
     - Frame 


 - **IPv4 Header Fields :**

     - **1- Version :**

         - 4-bits in length 
         * identfies the version of IP used ( IPv4 - IPv6 )


     * **2- Internet Header Length ( IHL ) :**

         - 4-bits in length 
         - indicates the total length of the header
         - identifies the length of the header in 4-byte increments  ( Maximum value = 15 (60 bytes) ) , ( Minimum Value = 5 (20 bytes))

     - **3- DSCP :**

         - Differentiated Sevices Code Point 
         - 6-bits in length 
         - used for QoS ( Quality of Service )  
         - used to prioritize delay-sensitive data 

     * **4- ECN :**

         - Explict Congestion Notification 
         - 2-bits in length 
         - provides end-to-end notification of network congestion without dropping packets 

     * **5- Total Length :**

         - 16-bits in length 
         - indicates the total length of the packet 

     * **6- Identification :**

         - if a packet is fragmented due to being to large , this field is used to idenify which packet the fragment belongs to

     * **7- Flags :**

         - 3-bits in length 
         - used to control / identify fragments  


     * **8- Fragment Offset :**

         - used to indicate the position of the fragment within the original , unfragmented IP packet 
         - allows fragmented packets to be reassembled even if the fragments arrive out of order 


     * **9- Time To Live :** 

         - 8-bits in length 
         - A Router will drop a packet with a TTL of 0 
         - used to prevent infinite loops 
         - originally designed to indicate the packet’s max lifetime in seconds 
         - in practice , indicates a " hop count " : each time the packet arrives at a router decreases the TTL by 1
         - recommended default TTL is 64

     * **10- Protocol :**

         - 8-bits in length 
         - indicates the protocol of the encaosulated layer 4 PDU ( TCP with value of 6 - UDP with value of 17 - ICMP with value of 1 - OSPF with value of 89 )  

     * **11- Header Checksum :**

         - 16-bits in length 
         - used to check for errors in the IPv4 header 


     * **12- Source/Destination IP address :**

         - 32-bits in length (each)
         - Source IP address = IPv4 address of the sender of the packet 
         - Destination IP address =  IPv4 address of the intended reciever of the packet     

     
     * **13- Options :**

         - length : 0-320 bits 
         - rarely used ( if IHL > 5 )

# **VLANs**


 * **LAN :** 

     * is a single broadcast domain, including all devices in that broadcast domain 

 - **Broadcast Domain :**

     - is the group of devices which will recieve a broadcast frame sent by any one of the members  .


 * The switch will not forward traffic between VLANs including broadcast / unknown unicast traffic 

 * The switch does not perform inter-VLAN routing it must send the traffic through the router 

 * **Access (untagged) port :**

     * is a switchport which belongs to a single VLAN and usually connects to end hosts 

 - **trunk (tagged) ports :**

     - carry traffic from multiple VLANs over a single interface .

     - switch will tag all frames that they send over a trunk link which allows the recieving switch to know which VLAN the frame belongs to 

     - **trunking protocols :**

         * **1- ISL (Inter-Switch Link) :**
           
             - probably never used in real world 

         - **2- IEEE 802.1Q :**

             - inserted between the source and type fields of the ethernet frame 

             - 4-bytes in length 

             - consists of two main fields :

                 * **1- Tag Protocol Identifier (TPID) :**

                     - 16-bits in length 
                     - always set to a value of 0x8100 which indicates that the frame is 802.1Q tagged
                       

                 * **2- Tag Control Information (TCI) :**

                     - has three subfields : 
                      
                         * **1- PCP (Priority Code Point) :**
                           
                             - 3-bits in length 
                             - used for class of service (CoS) which prioritize important traffic in congested networks 
                             
                         * **2- DEI (Drop Eligible Indicator) :**

                             - 1-bit in length 
                             - used to indicate frames that can be dropped if the network is congested . 

                         * **3- VID (VLAN ID) :**

                             - 12-bits in length 
                             - identifies the VLAN the frame belongs to 


 * **VLAN Ranges**

     - (1 - 4094) is devided into two sections :

         * **normal VLANs :**   1 - 1005
         * **extended VLANs :**  1006 - 4094 


 * **Native VLAN :**

     - is VLAN 1 by default on all trunk ports , can be manually configured on each trunk port 
     
     - the switch does not add an 802.1Q tag to frames in the native VLAN 
     
     - when the switch recieves an untagged frame on a trunk port it assumes the frame belongs to the native VLAN


 * **ROAS (Router on a Stick) :**


     - used to route between multiple VLANs using a single interface on the router and switch 

     - the switch interface is configured as aregular trunk 
     
     - the router interface is configured using subinterfaces 
     
     - the router will behave as if frames arriving with a cetain VLAN tag have arrived on the subinterface configured with that VLAn tag 
     
     - the router will tag frames sent out of each subinterface with the VLAN tag configured on the subinterface  


 
 * **Multilayer Switch**

     - capable of both switching and routing .


 - **SVIs (Switch Virtual Interfaces) :**

     * are the virtual interfaces you can assign IP addresses to in a multilayer switch 
         
     



                                   
         
           






# **DTP & VTP**



 * **DTP (Dynamic Trunking Protocol) :**

     - cisco proprietary protocol that allows Cisco switches to dynamically determine thier interface status (access or trunk) without manualconfiguration 

     - enabled by default on all Cisco switch interface 

     - To Manually configure switch port we use the following commands :

         * (switchport mode access -or- switchport mode trunk)

     - for security purpuses , DTP must be disabled 

     - A switchport in **dynamic desirable** mode will actively try to form a trunk with other Cisco switches 

     - A switchport in **dynamic auto** mode will not  actively try to form a trunk with other Cisco switches, however it will form a trunk if the switch connected to it is actively trying to form a trunk 

     - on older switches , **switchport mode dynamic desirable** is the default administrative mode 

     - on newer switches , **switchport mode dynamic auto** is the default administrative mode

     - To disable DTP negotiation we use this command ==> switchport nonegtiation or by configuring an access port with switchport mode access 

     - switches that support both **802.1Q** and **ISL** trkunk encapsulation can use DTP to negotiate the encapsulation they will use 

     - this negotiation is enabled by default , as the default trunk enapsulation mode is : **switchport trunk encapsulation negotiate**




 * **VTP (VLAN Trunking Protocol) :**

     - allows you to configure VLANs on a central VTP server , and other switches (VTP clients) will sync their VLAN database to the server 

     - There are three VTP versions :  ( 1 - 2 - 3 )

     - **VTP modes :**

         * **1- Server :**

             - can ( add - modify - delete ) VLANs
             
             - store the VLAN database in NVRAM ( VLAN database is saved even if the switch is turned off or reloaded )
             
             - will increase **the revision number** every time a VLAN is added/modified/deleted
             
             - **Revision Number :**

                 * it’s what VTP uses to determine the newest version of the VLAN database , the version that the switch will sync to

             - will advertise the latest version of the VLAN database on trunk interface , and the VTP clients will sync their VLAN database to it 

             - VTP server will sync to another VTP server with a higher revision number 

         
         * **2- Client :**

             * can not ( add - modify - delete ) VLANs

             * do not store the VLAN database in NVRAM ( in VTPv3 they do)

             * will sync their VLAN database to the server with the highest revision number in their VTP domain 

             * will advertise their VLAN database , and forward VTP advertisement to other clients over their trunk ports 


         * **3- Transparent :**

             - does not sync its VLAN database 

             - maintain its own VLAN database in NVRAM , can ( add - modify - delete ) VLANs , but they will not be advertised to other switches 

             - will forward VTP advertisements that are in the same domain as it 

             - changing the VTP domain to an unused domain will reset the revision num to 0

             - changing the VTP mode to transparent will also reset the revision num to 0

    

     * Cisco switches operate in VTP server mode by default 

     * if a switch with no VTP domain recieves a VTP advertisement with VTP domain name , it will automaticalyy join that VTP domain 

     * if a switch recieves a VTP advertisement in the same VTP domain with higher revision , it will update its VLAN database to match 

     * if you connect an old switch with a higher revsion number to your network and the VTP domain name matches , all switches in the domain will sync their VLAN database to that switch






# ***STP (Spanning Tree Protocol) 1***


 * **Before STP**
 
     * the Ethernet header does not have a TTL field , these broadcast frames will loop around the network indefinitely , if enough of these looped broadcasts accumulate in the network . the network will be too congested for legitimate traffic to use the network . this is called a **Broadcast Storm**


     * each time a frame arrives on a switchport , the switch uses the source MAC address field to learn the MAC address and update its MAC address table , when frames with the same source MAC address repeatedly arrive on different interfaces , the switch is continuously updating the interface in its MAC address table . this is known as **MAC Address Flapping**


 * STP prevents layer 2 loops by placing redundant ports in a blocking state , essentially disabling the interface 

 * these interfaces act as backups that can enter a forwarding state if an active interface fails 

 * interfaces in ablocking state only send or recieve STP messages called **BPDUs (Bridge Protocol Data Units)**

 * **How STP works :**

     - by selecting which ports are **forwarding** and which ports are **blocking**, STP creates a single path to/from each point in the network 

     - there are a set process that STP uses to determine which ports should be forwording and which should be blocking 

     - STP-enabled switches send/recieve Hello BPDUs out of all interfaces , the defualt timer (2 seconds)

     - if a switch recieves a Hello BPDU on an interface , it knows that interface is connected to another switch 


 * switches use one field in the STP BPDU , the **Bridge ID field**, to elect a **root bridge** for the network 

     * the switch with the lowest **Bridge ID** becomes the **root bridge**

     * all ports on the **root bridge** are put in a forwarding state , and other switches in the topology must have a path to reach the root bridge  

     * all interfaces on the root bridge are **designated ports** ( in a forwarding state )
  


 * **Bridge ID field :**

     - have two subfields :

         - **1- Bridge priority :**

             * 16-bits in length 
             * default bridge priotiry is 32768 ( actually 32769 ==> 32768 + 1 ) on all switches , so by default the MAC address is used as the tie-breaker 
             
             * bridge priority has been updated to be made of two parts :

                 * 1- bridge priority (4 bits)
                 * 2- VLAN ID (12 bits) :

                     - included because Cisco switches use a version of STP called **PVST (Per-VLAN Spanning Tree)** . PVST runs a separate STP instance in each VLAN , so in each VLAN different interfaces can be forwarding/blocking 
                     
                     - by adding the VLAN ID into the bridge priority , the switch will have a different Bridge ID in each VLAN 


         - **2- MAC address** 

             * 48-bits in length   



 * **Root cost :**

     - is the total cost of the outgoing interfaces along the path to the root bridge  

 - the ports connected to another switch’s root port must be designated . because the root port is the switch’s path to the root bridge , another switch must not block it 




 * >> **STP Proccess Summary**

     * 1- one switch is elected as the **root bridge** . all ports on the root bridge are **designated ports** (in forwarding state) .

     Root bridge selection :

         - a ) lowest bridge ID
         - b ) lowest MAC address

     * 2- each remaining switch will select one of its interfaces to be its **root port** . ports across from the root port are always designated ports.

     Root port selection :

         - a ) lowest root cost 
         - b ) lowest neighbor bridge ID
         - c ) lowest neighbor port ID

     * 3- each remaining collision domain will select one interface to be a designated port . the other port in the coolision domain will be non-designated (blocking) 

     Designated port selection :

         - a ) interface on the switch with lowest root cost 
         - b ) interface on switch with lowest bridge bridge ID           



 
 

# ***STP (Spanning Tree Protocol) 2***



 * **STP port states :** 
    
     - **1-Blocking (Stable) :**

         * non-designated ports are in a blocking state 
         
         * interfaces in a blocking state are effectively disable to prevent loops 
         
         * interfaces in a blocking state do not ( send/recieve regular network traffic - forward STP BPDUs - learn MAC addresses )
         
         * interfaces in a blocking state recieve STP BPDUs 

     - **2- Listening (Transitional) :**

         * only designated or root ports enter the listening state 

         * the listening state is 15 seconds long by default . this is determined by **forward delay** timer   

         * interface in the listening state only forwards/recieves STP BPDUs

         * interface in the listening state does not ( send/recieve regular traffic  -  learn MAC addresses from regular traffic that arrives on the interface )   


     - **3- Learning (Transitional) :**


         * the learning state is 15 seconds long by default . this is determined by **forward delay** timer   

         * interface in the learning state only forwards/recieves STP BPDUs 

         * interface in the learning state does not send/recieve regular traffic 

         * interface in the learning state learns MAC addresses from regular traffic that arrives on the interface 


     - **4- Forwarding (Stable) :**


         * a port on the forwarding state sends/recieves ( BPDUs and normal traffic ) 

         * a port on the forwarding state learns MAC addresses .


     - **5- Disabled**  

 * * * 

 * **STP timers :**


     * **Hello :**

         - duration ==> 2 sec
         - how often the root bridge sends hello BPDUs .

     * **Forward delay :**

         - duration ==> 15 sec
         - how long the switch will stay in the listening and learning (each state is 15 seconds)

     * **Max Age :**

         - duration ==>  10 * Hello
         - how long an interface will wait after ceasing to recieve Hello BPDUs to change the STP topology 

 * * * 

 * PVST ==> only ISL trunk encapsulation
 * PVST+ ==> supports 802.1Q

 * * *


 * >> **STP Optional features (Toolkit) :**

   
     - **1- Portfast :**

         * can be enabled on interfaces which are connected to end hosts 

         * allows a port to move immediately to the forwarding state , bypassing listening and learning 

         * enbaled by commands : 

             - SW(config-if)# spanning-tree portfast
         
             - SW(config)# spanning-tree portfast default

         * portfast can also cause loops if the network cabling is changed without proper caution 


     - **2- BPDU Guard :**

         - if an interface with BPDU guard enabled recieves a BPDU from another switch , the interface will be shut down to prevent a loop from forming   

         - To configure it :

             - SW(config-if)# spanning-tree bpduguard enable
        
             - SW(config)# spanning-tree portfast bpduguard default 


     - **3- Root Guard :**

 Therefore, RSTP offers am improvment over the 30 seconds or more that 802.1D takes to move a link to forwarding .
 the heart of the protocl is a new bridge handshack mechanism, which allows ports to move directly to forwarding

# ***Dynamic Routing***


 * **Network route :**

     - a route to a network/subnet (mask length < /32)

 - **Host route :**

     - a route to a specific host (/32 mask)    

 * * *

 * routers can use dynamic routing protocols to advertise information about the routes they know to other routers  

 * the form adjacencies with adjacent routers to exchange this info 

 * if multiple routes to a destination are learned , the router determines which route is superior and adds it to the routing table . it uses the metric of the route to decide which is superior 


 * * *

 * **Types of Dynamic Routing protocols :**

     - **1- IGP (Interior Gateway Protocol) :**

         - used to share routes within a single autonomous system (AS) , which is a single organization

         - cataegorized by **algorithm type** (the proccess used by each protocol to share route info and determine the best route to each dst) into :

             * a ) **Distance vector :**
              
                 - 1- RIP (Routing Information Protocol)

                 - 2- EIGRP (Enhanced interior Gateway Routing Protocol) :


             * b ) **Link State :**

                 - 1- OSPF (Open Shortest Path First) :
            
                 - 2- IS-IS(intermediate sysytem to intermediate system) :


     - **2- EGP (EXterior Gateway Protocol) :**

         - used to share routes between different autonomous systems 

         - algorithm type :

             * path vector :

                 - BGP (Border Gateway Protocol)


 * * *

 * **Distance Vector Routing Protocols**


     * called distance vector because the routers only learn the distance (metric) and vecor (direction, the next hop router) of each router 

     * distance vector protocols operate by sending the following to their directly connected neighbors :

         - their known destination networks 
         - their metric to reach their known destination networks 

     - this method of sharing route information is often called **routing by rumor**


 * * *


 * **Link State Routing Protocols**


     * every router creates a " connectivity map " of the network 

     * to allow this, each router advertise information about its connected networks to its neighbors . these advertisements are passed along to other routers, until all routers in the network develop the same map of the network 

     * each router independently uses this map to calculate the best routes to each destination 

     * link state protocols use more resources (CPU) on the router , because more information is shared 

     * link state protocols tend to be faster in reacting to changes in the network than distance vector protocols 


 * * *


 * **Dynamic Routing protocols metrics**


     * routers determine which is the best route using the **metric value** (lower metric = better route) 

     * if a router leaarns two or more routes via the same routing protocol to the same destination with same metric, both will be added to the routing table . the traffic will be load-balanced over both routes ==> ECMP (Equal Cost Multi-Path)       

     * static routes do not use the concept of metric 

     * **Administrative Distance (AD) :**

         - used to determine which routing protocol is prefered 

         - a lower AD is prefered , and indicates that the routing protocol is considered more trustworthy (more likely to select good routes)

# ***RIP / EIGRP***


 * **1- RIP (Routing Information Protocol) :**

     * dinstance vector IGP (uses routing-by-rumor logic to learn and share routes)

     * uses hop count as its metric (one router = one hop) 

     * the maximum hop count is 15

     * has three versions :

         * **1- RIPv1 :**

             * only advertise classful addresses 
             * does not support VLSM, CIDR
             * does not include sunten mask information in advertisement (response message)
             * messages are broadcast to 255.255.255.255 

         * **2- RIPv2 :**

             * supports VLSM, CIDR
             * includes subnet mask information in advertisement 
             * messages are multicast to 244.0.0.9

         * **3- RIPng (RIP Next Generation)**

     * uses two messages types :

         - **1- Request :**

             - to ask RIP-enabled neighbor routers to send their routing table 

         - **2- Response :**

             - to send the local router’s routing table to neighboring routers

     * by default , RIP-enable routers will share their routing table every 30 seconds  


  * * *

 * **2- EIGRP (Enhanced Interior Gateway Routing Protocol) :**

     * considered an advanced/hybrid distance vector routing protocol 
     * faster than RIP in reacting to changes in the network 
     * does not have the 15 hop-count limit of RIP
     * sends messages using multicast address 244.0.0.10
     * is the only IGP that can perform unequal-cost load-balancing (by default it performs ECMP load-balancing over 4 paths like RIP )

     * uses a **wildcard** mask instead of a regular subnet mask 

         * **wildcard mask :**

             - is basically an inverted subnet mask 
             - all 1s in the subnet mask are 0 in the equivalent wildcard mask .all 0s in the subnet mask are 1 in the equivalent wildcard mask



# ***OSPF***


 * uses the **SPF (Shortest Path First)** algorithm . also callled **Dijkstra’s** algorithm 

 * have three versions :

     * 1- OSPFv1 
     * 2- OSPFv2 : used for IPv4
     * 3- OSPFv3 : used for IPv4 and IPv6

 * routers store information about the network in **LSAs (Link State Adevertisements)** , which are organized in a structure called the **LSDB (Link State Database)**    

 * routers will flood LSAs until all routers in the OSPF area develop the same map of the network (LSDB)

 * each LSA has an aging timer (30 min by default) . the LSA will be flooded again after the timer expires 

 
 * * * 

 * **main 3 steps in the process of sharing LSAs and determining the best route :** 

   
     * 1- **become neighbors** with other routers connected to the same segments .
     
     * 2- **exchange LSAs** with neighbor routers 

     * 3- **calculate the best routes** to each destination , and insert them into the routing table  


 * * *

 * **OSPF Areas :**

     * OSPF uses **areas** to divide up the network 

     * small networks can be single-area without any negative effects on performance 

     * in larger networks, a single-area design can have negative effects :

         - the SPF algorithm takes more time to calculate routes 
         - the SPF algorithm requires exponentially more peocessing power on the routers 
         - the larger LSDB takes more memory on the routers 
         - any small change in the network causes every router to flood LSAs and run the SPF algorithm again


     * by dividing a large OSPF network into several smaller areas , you can avoid the above negative effects 


     * **Area :** 

         - is a set of routers and links that share the same LSDB

     * **the backbone area (area 0) :**

         - is an area that all other areas must connect to 

     * **internal routers :**

         - routers with all interfaces in the same area

     * **area border routers (ABRs) :**

         - routers with interfaces in multiple areas 
         - ABRs maintain a separate LSDB for each area they are connected to .
         - it is recommended that you connect an ABR to a maximum of 2 areas     

     * **intra-area route :**

         - is a route to a destination inside the same OSPF area

     * **intraarea route :**

         - is a route to a destination in a different OSPF area 

     * **autonomous system boundary router (ASBR) :**

         - is an OSPF router that connects the OSPF network to an external network 



 * * *

 * **OSPF areas Rules**

  - OSPF areas should be contiguous
  - all OSPF must have at least on ABR connected to the backbone area
  - OSPF interfaces in the same subnet must be in the same area        

 * * *

 * **OSPF Metric (Cost) :**

     - the OSPF cost to a destination is the total cost of the outgoing interfaces (loopback interfaces have a cost of 1) 

     - it is auotmatically calculated based on the bandwidth (speed) of the interface
     
     - it is calculated by dividing a **reference bandwidth** value by the interface’s bandwidth 

     - the default reference bandwidth is 100 mbps

     - FastEthernet, Gigabit Ethernet, 10Gig Ethernet are equal and all have a cost of 1 by default 

     - all values less than 1 will be converted to 1

     - you should change the reference bandwidth with this command :

         * R1(config-router)# **auto-cost** **reference-bandwidth** mbps .


     * **three ways to modify the OSPF cost :** 

         - 1- change the reference bandwidth :

             * R1(config-router)# **auto-cost** **reference-bandwidth** mbps   

         - 2- manual configuration :

             * R1(config-if)# **ip ospf cost** cost

         - 3- change the interface bandwidth :

             * R1(config-if)# **bandwidth** kilobits-per-second  


 * * *


 * **OSPF Neighbors :**

     * once routers become neighbors, they automatically do the work of sharing network info 

     * when OSPF is activated on an interface , the routers starts sending OSPF **hello** messages out of the interface at regular intervals (determined by the hello timer) . these are used to introduce the router to potential OSPF neighbors 

     * the default hello timer is 10 seconds

     * hello messages are multicasr to 244.0.0.5 

     * OSPF messages are encapsulaated in an IP header , with a value of 89 in the protocol field 



  * **OSPF Neighbors states :**

     - 1- **Down :**

         * when the OSPF neighbor is unkmown 

     - 2- **init :**

         * hello packet recieved , but own router ID is not in the hello packet 

     - 3- **2-way state :**    

         * the router has recieved a hello packet with its RID in it

         * if both routers reach the 2-way state, it means that all of the conditions have been met for them to become OSPF neighbors . they now ready to share LSAs to build a common LSDB

         * in some network types, a DR (Designated Router) and BDR (Backup Designated Router) will be elected at this point  

     - 4- **Exstart :**

         * before exchanging , the two router will choose which one will start the exchange 

         * the router with the higher RID will become the **Master** and initiate the exchange . the router with the lower RID will become the **Slave**

         * to decide the master and the slave , they exchange DBD (Database Description) packets 


     - 5- **Exchange :**

         * the routers exchange DBDs which contain a list of the LSAs in their LSDB

         * these DBDs do not include detailed information about LSAs

         * the routers compare the information in the DBD they recieved to the information in their own LSDB to determine which LSAs they must recieve from their neighbor 


     - 6- **Loading :**

         * routers send Link State Request (LSR) messages to request that their neighbor send them any LSAs they do not have 

         * LSAs are sent in Link State Update (LSU) messages 

         * the routers send LSAck messages to achknowledge that they recieved the LSAs 


     - 7- **Full :**

         * the routers have a full OSPF adjacency and identical LSDBs 

         * they continue to send and lestin for hell packets (every 10 seconds by default ) to maintain neighbor adjacency 

         * every time a hello packet is recieved , the dead timer (40 seconds by default) is reset 

         * the routers will continue to share LSAs as the network changes to make sure each router has a complete and accurate map of the network (LSDB)


* * * 


 * **OSPF Network Types**

     - **1- Broadcast :**

         - enabled be default on **Ethernet** and **FDDI** (Fiber Fistributed Data Interface)
         
         - routers dynamically discover neighbor be sending/lestining for OSPF hell messages using multicast address 244.0.0.5

         - a DR (designated router) and BDR (backup designated router) must be elected on each subnet 

         - routers which are not the DR or BDR become **DROther**

         - the DR/BDR election order of priority :

             1- highest **OSPF interface priority**
             2- highest OSPF RID

         - first place becomes the DR for the subnet , second place becomes the BDR

         - if you set the OSPF interface priority to 0 , the router can not be the DR/BDR for the subnet 

         - the DR/BDR election is non-preemptive . oncce the DR/BDR are selected they will keep their role until OSPF is reset , the interface fails/is shut down

         - when the DR goes down, the BDR becomes the new DR . then an election is held for the next BDR  
         
         - routers will only form a full OSPF adjacency with the DR and BDR of the segment . therefore, routers only exchange LSAs with the DR and BDR . DROther will not exchange LSAs with each other 

         - all routers will still have the same LSDB , but this reduces the amount of LSAs flooding the network

         - messages to the DR/BDR are multicast using address 244.0.0.6






     * **2- Point-to-point :**

         - enabled by default on serial interfaces using the **PPP** (point-to-point protocool) and **HDLC** (High-Level Data Link Control) encapsulation 

         - routers dynamically discover neighbor be sending/lestining for OSPF hell messages using multicast address 244.0.0.5

         - a DR and BDR are not elected
         - these encapsulation are used for point-to-point connections 
         - the two routers will form a full aadjacency with each other 

         - **Serial interfaces :**

             * one side of a serial connection functions as DCE (Data Communication Equipment)
             * the other side functions as DTE (Data Terminal Equipment) 
             * the DCE side needs to specify the clock rate (speed) of the coonection  

         * you can conigure PPP encapsulation with this command :

             * R1(config-if)# **encapsulation ppp**

         * identify which side is DCE/DTE :

             * R1# **show controllers** interface-id

         * you must configure the clock rate on the DCE side :

             - R1(config-if)# **clock rate** bits-per-seconds            

     - **3- Non-broadcast :**

         - enabled by default on **frame relay** and **X.25** interfaces    


 * * *


 * **OSPF Neighbor Requirements :**

     * 1- area number must match
     * 2- interfaces must be in the same subnet 
     * 3- OSPF process must not be shutdown
     * 4- OSPF router IDs must be unique 
     * 5- hello and dead timers must match
     * 6- authentication settings must match
     * 7- IP MTU settings must match
     * 8- OSPF network type must match   


 * 7 and 8 not neccasary ( can become OSPF neighbors , but OSPF does not operate properly )

# ***FHRPs (First Hop Redundancy Protocols) :***


 * **FHRP :**
   
     - a computer networking protocol which is designed to protect the default gateway used on a subnetwork by allowing two or more routers to provide backup fot that addres ; in the event of failure of an active router , the backup router will take over the address , usually within a few seconds 

     - a vitual IP is configured on the two routers , and a virtual MAC is generated for the virtual IP ( each FHRP uses a different format for the virtual MAC ) 

     - an **active** router and a **standby** router is elected . ( different FHRPs use different terms )

     - end hosts in the network are configured to use the virtual IP as their default gateway 

     - the active router replies to ARP requests using the virtual MAC address , so traffic destined for other networks will be sent to it 

     - if the active router fails , the standby becomes the next active router . the new active router will send **gratuitous ARP** messages so that switches will update their MAC address tables . it now functions as the default gateway 

     * **gratuitous ARP :**
      
         - ARP replies sent without being requested

     * if the old active router comes back online , by default it will not take back its role as the active router . it will become the standby router 

     * you can configure **preemption** so that the old active takes back its old role 

 
 * * *

 * **HSRP (HOT Standby Router Protocol) :**

     - an **active** and **standby** router are elected

     - have two versions : Version 1 and Version 2

         * version 2 adds IPv6 support and increases the number of groups that can be configured 

     * multicast IPv4 address :

         * v1 = 224.0.0.2
         * v2 = 224.0.0.102

     * virtual MAC address :

         * v1 = 0000.0c07.ac**XX** (XX = HSRP group num)  
         * v2 = 0000.0c9f.f**XXX**

     * in a situation with multiple subnets/VLANs , tou can configure a different active router in each subnet/VLAN to load balance   


 * * *


 * **VRRP (Virtual Router Redundancy Protocol) :** 

     * a **master** and **backup** router are elected 

     * multicast IPv4 address :   244.0.0.18

     * virtual MAC address : 0000.5e00.01**XX** ( XX = VRRP group num )

     * in a situation with multiple subnets/VLANs , you can configure a different master router in each subnet/VLAN to load balance 


 * * *

 * **GLBP (Gateway Load Balancing Protocol) :** 

     - load balances among multiple routers within a single subnet 

     - an AVG (Active Virtual Gateway ) is elected 

     - up to four AVFs (Active Virtual Forwarders) are assigned by the AVG (the AVG itself can be an AVF)

     - each AVF acts as the default gateway for a portion of the hosts in the subnet 

     - multicast IPv4 address :  224.0.0.102

     - virtual MAC address :

         - 0007.b400.**XXYY**  ( XX = GLBP group num , YY = AVF num )
         
           


# ***TCP & UDP***


 * **functions of layer 4 (Transport Layer) :**

     * provides transparent transfer of data between end hosts 

     * provides (or not) various services to apps :

         - reliable data transfer 
         - error recovery 
         - data sequencing 
         - flow control 

     - provides layer 4 addressing (port numbers) .

         * identify the application layer protocol 
         * provides session multiplexing 

             - **Session :**

                 * is an exchange of data between two or more communicating devices

         * port number ranges that have been designated by IANA (Internet Assigned Numbers Authority) :

             * 1- **well-known** port numbers :   0 - 1023
             * 2- **Registered** port numbers :  1024 - 49151
             * 3- **(Ephemeral/private/dynamic)** port numbers :  49152 - 65535

* * *
* * *          


 * **TCP (Transmission Control Protocol) :**

     - **is connection-oriented** 

         * before actually sending data to the destination host , the two hosts communicate to establish a connection . once the connection is established , the data exchange begins

     * **provides reliable communication**

         - the destination host must acknowledge that it recieved each TCP segment 

         - if a segment is not acknowledged , it is sent again

     - **provides sequencing**

         * sequence numbers in the  TCP header allow destination hosts to put segments in the coorect order even if they arrive out of order 

     * **provides flow control**

         - the destination host can tell the source host to increase/decrease the rate that data is sent  

 * * *

 * **Establishing connections (Three-Way Handshake) :**

     - 1- src host sends **SYN** flag to dst host 
     - 2- dst host sends **SYN , ACK** flags to src host
     - 3- src host sends **ACK** flag to dst host  

     * after the handshake the real data exchange can begin          


 * **Terminating connections (Four-Way Handshake) :**

     - 1- src host sends **FIN** flag to dst host 
     - 2- dst host send **ACk** flag to src host 
     - 3- dst host send **FIN** flag to src host

* * *

 * **TCP : (Sequencing / Acknowledgement) :**

     - hosts set a random initial sequence number         
     
     - **forward acknowlegement** is used to indicate the sequence number of the next segment the host expects to recieve    

   
 * **TCP Retransmission :**

     - 1- src sends dst a segment with sequence num1 (using sequencing)
     
     - 2- dst sends ACK num1 to src (using forward acknowlegement)
     
     - 3- src sends sequence num2 , but for some reason it does not reach dst
     
     - 4- after waiting a certain amount of time with no ACK from dst , src resends the segment       

* * *

 * **TCP flow control : (window size)**

     - the TCP header’s **window size** field allows more data to be sent before an acknowledgement is required

     - for example : a host could send 3 sements with sequnce nums 1, 2, and 3 and then an ACK is sent 

     - a **sliding window** can be used to dynamically adjust how large the window size is 

* * *

 * **Important port numbers that use TCP :**

     - 1- FTP data (File Transfer Protocol) (20)
     - 2- FTP control (21)
     - 3- SSH (Secure Shell) (22)
     - 4- Telnet (Teletype Network) (23)
     - 5- SMTP (Simple Mail Transfer Protocol) (25)
     - 6- HTTP (HyperText Transfer Protocol) (80)
     - 7- HTTPS (HyperText Transfer Protocol Secure) (443)
     - 8- POP3 (Post Office Protocol version 3) (110)
     - 9- DNS (Domain Name System) (53)  


 * * *
 * * *


 * **UDP (User Datagram Protocol) :**

     - not connection oriented 

     - does not provide reliable communication , segments are sent best-effort

     - does not provide sequencing

     - does not provide flow control 


     * **Important port numbers that use UDP :**

         - 1- DHCP (Dynamic Host Control Protocol) server (67)
         - 2- DHCP client (68)
         - 3- TFTP (Trivial File Transfer Protocol) (69)
         - 4- SNMP (Simple Network Management Protocol) agent (161)
         - 5- SNMP manger (162)

# ***IPv6***


 * **Why IPv6 ???**

     - the main reason is that there simply are not enough IPv4 address available


* * *

 * An IPv6 address is **128-bits (8 bytes)**

 * written as 32 haxadecimal characters , divided into 8 groups of 4 using colons (:) 

* * *

 * **shortening (abbreviation) IPv6 addresses:**

     - leading 0s can be removed 
     
     - consecutive quartets of all 0s can be replaced with a double colon :: , **(can only be used once)**

         * ex :2001:0db8:0000:0000:0000:0000:0080:562a   ==> 2001:0db8::0080:562a  ==> 2001:0db8::80:562a

     * if there are two equal-length choices for the :: , use :: to shoten the one on the left  

         * ex :  2001:0db8:0000:0000:0f2a:0000:0000:00ba   === >    2001:db8::f2a:0:0:bba     

     - hexadecimal characters must be written using lower-case    

* * * 

 * **fining the IPv6 prefix (global unicast addresses) :**

     - typically , an enterprise requesting IPv6 addresses from ISP will recieve a a/48 block 

     - typically , IPv6 subnets use a  /64 prefix length

     - that means an enterprise has 16 bits to use to make subnets 

     - the remaining 64 bits can be used for hosts 

     - **EX :**

         * 2001:0db8:8b00:0001:0000:0000:0000:0001/64

             - (2001:db8:8B00)  : 4-bit **global routing prefix**, assigned by the ISP 

             - (0001) : 16-bit **subnet identifier**, used by the enterprise to make various subnets 

             - (0000:0000:0000:0001) 64-bit **interface identifier**, the host portion of the address 


     * **finding IPv6 prefix (Network address) examples :**

         - 300d:00f2:0b34:2100:0000:0000:1200:0001/56  ===>  300d:F2:b34:2100::/56

         - 2001:0db8:8B00:0001:fb89:017B:0020:0011/93   ===>   2001:0db8:8b00:1:fb89:178::/93     



* * *  


* **configuring IPv6 addresses :**

     * to perform IPv6 routing :

         - R1(config)# ipv6 unicast-routing 

     * to configure an IPv6 address :

         - R1(config-if)# **ipv6 address** the address and prefix

             - ex ==> R1(config-if)# ipv6 address 2001:db8:0:0:1/64


* * *

* **configuring IPv6 addresses (EUI-64):**

 - EUI (Extended Unique Identifier)

 - it is a method of converting a MAC address(48 bits) into a 64-bit interface identifier

 - this interface identifier can be then become the host portion of a  /64 IPv6 address

 - **How to convert the MAC address ?**

     - 1- divid the MAC address in half :

         - 1234 5678 90AB ===>  1234 56 **|** 78 90AB

     - 2- insert **FFFE** in the middle :

         - 1234 56FF FE78 90AB

     - 3- invert the 7th bit (0=>1 , 1=>0) :        

         - 1234 56FF FE78 90AB  ===> 1034 56FF FE78 90AB


 - **to configure an interface using EUI-64 :**

     - R1(config-if)# **ipv6 address** network prefix **eui-64**

         - **Ex :** 

             * R1(config-if)# ipv6 address 2001:db8::/64 eui-64


 * **why invert the 7th bit ?**

     - MAC addresses cab be divided into two types :

         - 1- **UAA (Universally Administered Address) :**
            
             * uniquely assigned to the device by the manufacturer

         - 2- **LAA (Locally Administered Address) :**

             * manually assigned by an admin (with the **mac-address** command on the interface) or protocol . does not have to be globally unique 

     - you can identify a UAA or LAA by the 7th bit of the MAC address , called the U/L bit 

         - U/L bit set to 0 = UAA
         - U/L bit set to 1 = LAA


     - in the context of IPv6 addresses/EUI-64 , the meaning of the U/L bit is reversed 

         - U/L bit set to 1 = the MAC address the EUI-64 interface ID was made from was an LAA

         - U/L bit set to 0 = the MAC address the EUI-64 interface ID was made from was an UAA

 * **Gloabal unicast addresses :**

     - public addresses which can be used over the internet 

     - must register touse them . because they are public addresses , it is expected that they are globally unique 

     - originally defined as the **2000::/3** block (2000::  to  3fff:ffff:ffff:ffff:ffff:ffff:ffff:ffff) 


     * **EX ==>**  2001:0db8:8b00:0001:0000:0000:0000:0001/64

         - (2001:0db8:8b00)  : 4-bit **global routing prefix**, assigned by the ISP 

         - (0001) : 16-bit **subnet identifier**, used by the enterprise to make various subnets 

         - (0000:0000:0000:0001) 64-bit **interface identifier**, the host portion of the address  


* * *

 * **Unique local addresses :**

     - private addresses which can not be used over the internet 

     - you do not need to register to use them . they can be used freely within internal networks and do not need to be globally unique . can not routed over the internet 

     - uses the address block **fc00::/7** (fc00::  to  fdff:ffff:ffff:ffff:ffff:ffff:ffff:ffff)

     - however , a later update requires the 8th bit to be sit to 1 , so the first two digits must be FD

     - the global ID should be so that addresses do not overlap when companies merge 


     * **EX ==>** fd45:93ac:8a8f:0001:0000:0000:0000:0001/64

         - fd  : indicates a unique local address 

         - 45:93ac:8a8f  : 40-bit **global ID** which should be randomly generated 

         - 0001 : 16-bit **subnet identifier**, used by the enterprise to make various subnets 

         - (0000:0000:0000:0001) 64-bit **interface identifier**, the host portion of the address

 * **Link local addresses :**


     - automatically generated on IPv6-enabled interface 

     - use command R1(config-if)# **ipv6 enable**  on an interface to enable ipv6 on an interface 

     - uses the address block **fe80::/10**  (fe80::  to  febf:ffff:ffff:ffff:ffff:ffff:ffff:ffff)

     - however, the standard states that the 54bits after fe80/10 should be all 0 , so you will not see link local addresses beginning with fe9, fea, feb . onle **fe8**

     - the interface ID is generated using EUI-64 rules 

     - link-local means that these addresses are used for communication within a single link (subnet) . Routers will not route packets with a link-local destination ipv6 address

     - **common uses of link-local addresses :**

         - routing protocols peerings (OSPFv3 uses link-local addresses for neighbor adjacencies)

         - next-hop addresses foe static routes 

         - Neighnor Discovery Protocol (NDP) ==> IPv6’s repalcement for ARP , uses link-local addresses to function 



* * *

 * **Multicast addresses :**

     - multicast addresses are one-to-many (one source to multiple destinations) that have joined the specific multicast group 

     - IPv6 uses range ff00::/8 for multicast (ff00::  to  ffff:ffff:ffff:ffff:ffff:ffff:ffff:ffff )

     - IPv6 does not use broadcast , but there is way to send a message to all hosts in the subnet using multicast  :

         - **ff02::1** ==> functions like broadcasr (sends a messages to all nodes/hosts)

         - **ff02::2** ==> sends a messages to all routers

         - **ff02::5** ==> sends a messages to all OSPF routers 

         - **ff02::6** ==> sends a messages to all OSPF DRs/BDRs

         - **ff02::9** ==> sends a messages to all RIP routers 

         - **ff02::A** ==> sends a messages to all EIGRP routers 


     - IPv6 defines multiple multicast **scopes** which indicates how far the packet should be forwarded 

         - **IPv6 multicast scopes :**

             * **1- interface-local (ff01) :**

                 - the packet does not leave the local device  
                 - can be used to send traffic to a service within the local device 

             - **2- link-local (ff02) :**

                 - the packet remains in the local subnet 
                 - routers will not route the packet between subnets 

             * **3- site-local (ff05) :**

                 * the packet can be forwarded by routers 
                 * should be kimited to a single physical location (not forwarded over a WAN)

             * **4- organization-local (ff08) :**

                 - wider in scope than site-local (an entire company/organization)

             - **5- Global (ff0e) :**

                 - has no boundaries . possible to be routed over the internet 


* * * 

 * **Anycast addresses :**

     - is one-to-one-of-many 

     - multiple routers are configured with the same IPv6 address 

         * they use a routing protocol to advertise the address 

         * when hosts sends packets to that destination address, routers will forward it to the nearest router configured with tha IP address (based on routing metric)


     * there is no specific address range for anycast addresses 

     * we can use a regular unicast address (global unicast , unique local) and specify it as an anycast address :

         - R1(config-if)# **ipv6 address 2001:db8:1:1::9/128 anycast**


* * *

 
 * **:: ==> the unspecified IPv6 address :**

     - can be used when a device does not yet know its IPv6 address 

     - IPv6 default routes are configured to  **::/0**


 - **::1 ==> the loopback address :**

     * used to test the protocol stack on the local device 

     * messages sent to this address are processed within the local device , but not sent to other devices 


* * *  


 * **IPv6 Header :**

     - **1- Version :**

         * 4-bits in length 
         * indicates the version of the IP that is used
         * fixed value of 6 (0b0110) to indicates IPv6

     - **2- Traffic Class :**

         * length : 8 bits 
         * used for Quality of Service (QoS) , to indicate high-priority traffic

     * **3- Flow Label :**

         - length : 20 bits
         - used to idenitfy specific traffic "flow" (communication between a specific src and dst)

     - **4- Payload Length :**

         * length : 16 bits
         * indicates the length of the payload (the encapsulated layer 4 segment) in bytes
         * the length of the IPv6 header itself is not included , bexause it is always 40 bytes


     * **5- Next Header :**

         - length : 8 bits 
         - indicates the type of the next header (header of the encapsulated segment) , for ex TCP or UDP
         - same function as the IPv4 header’s protocol field 

     - **6- Hop Limit :**

         * length : 8 bits 
         * the value in this field is decremented by 1 by each router that forwards it 
         * if it reaches 0, the packet is discarded 

     * **7- Src/Dst :**

         - length : 128 bits each
         - these fields contain the IPv6 addresses of the packet’s Src and the packet’s intended Dst     


* * *

 * **Solicited-Node Multicast Address :**

     * an IPv6 solicited-node multicast is calculated from a unicast address 

         - (ff02:0000:0000:0000:0000:0001:ff) + last 6 hex digita of unicast address

         - **ex ==>**  unicast address of 2001:0db8:0000:0001:0f2a:4fff:fea3:00b1 , find solicited-node multicast address 

             - ff02::1:ffa3:b1

 * **Neighbor Discovery Protocol (NDP) :**

     - replace ARP in IPv6
     
     - uses ICMPv6 and solicited-node multicast addresses to learn the MAC address of other hosts 

     - two message types are use :

         * **1- Neighbor Solicitation (NS) ==> ICMPv6 type 135 :**

         * **2- Neighbor Advertisement (NA) ==> ICMPv6 type 136**

     * another function of NDP allows hosts to automatically discover routers on the local network 

         * two messages are used for this process :

             * **1- Router Solicitation (RA) ==> ICMPv6 type 133 :**

                 - sent to multicast address ff02::2 
                
                 - asks all routers on the local link to identify themselves 
                 
                 - sent when an interface is enabled/host is connected to the network  

             - **2- Router Advertisment ==> ICMPv6 type 134 :**


                 * sent to multicast address ff02::1

                 * the router announce its presence , as well as other information about the link 

                 * these meassages are sent in response to RS messages   

                 * they are also sent periodically , even if the router has not recieved an RS    


* * *


 * **Stateless Address Auto-configuration (SLAAC) :**


     - hosts use RS/RA messages to learn the IPv6 prefix of the local link , and then automatically generate an IPv6 address

     - using ( **ipv6 address** prefix/prefix length **eui-64** ) command , you need to manually enter the prefix

     - using (**ipv6 address autoconfig**) command , you do not need to enter the prefix . the device uses NDP to learn the prefix used on the local link

     - the devic will use EUI-64 to generate the interface ID , or it will be randomly generated 
     
* * *

 * **Duplicate Address Detection (DAD) :**

     * allows host to check if other devices on the local link are using the same IPv6 address 

     - any time an IPv6-enable interface initializes (**no shutdown** command) , or an IPv6 address is configured on an interface , it performs DAD

     - DAD uses two messages :

         * NS  and  Na

     * the host will send an NS to its own IPv6 address . if it  does not get a reply , it knows the address is unique 

     * if it gets a reply , it means another host on the network is already using the address

     - a **connected network** route is automatically added for each connected network

     - a **local host** route is automatically added for each address configured on the router 

     - Routes for link-local addresses are not added to routing table

# ***ACL (Access Control List)***


 - security mechanism used in computer systems and networks to control and manage access to resources such as files, directories, or network resources. 
 
 - ACLs define which users or system processes are granted access to specific objects and what actions (read, write, execute, etc.) they are allowed to perform on those objects.


* * *

 * **How ACLs work**

     - ACLs in routers consist of a list of rules or entries that define which network packets are allowed or denied based on specific criteria, such as source IP address, destination IP address, port numbers, and protocol types

     - When a network packet arrives at the router, the router examines the packet headers to determine whether it matches any of the ACL rules.

     - ACL rules are evaluated in order, from the top of the list to the bottom. The router compares the packet's characteristics with the conditions specified in each rule.

     - Routers can have both inbound and outbound ACLs :

         * **Inbound ACLs :** are applied to packets entering the router, typically at the interface where the packets arrive

         * **Outbound ACLs :** are applied to packets leaving the router, usually at the interface where they exit.

     - **Implicit Deny:** If none of the rules in the ACL match the packet, the "ACL implicit deny" rule comes into play. This default rule denies the packet or request because no rule has explicitly permitted it.    

* * *


 * **ACL Types :**

     * **1- Standard ACLs:**

         - Standard ACLs permit or deny traffic based solely on the source IP address of packets.
         
         - These ACLs are often used for basic packet filtering and are generally less granular in their control.  

         - has two tpyes :

             - a ) **Standard named :**
                
                     * ACLs are assigned user-defined names to identify them

             - b ) **Standard numbered :**

                     *  assigned a numeric identifier from the range 1 to 99 or 1300 to 1999, depending on the networking device and software used. The number identifies the ACL.

         - should be applied as close to the destination as possible             

     * **2- Extended ACLs:**

         - Extended ACLs are more feature-rich and versatile compared to standard ACLs.

         - They consider various factors, including source and destination IP addresses, port numbers, protocols, and more.

         - Extended ACLs allow for more precise control over network traffic by specifying not only the source but also the destination of the traffic and the specific services or ports involved. 

         - have two types :

             * a ) **Extended named ACLs:**

                     - Similar to standard named ACLs, extended named ACLs are assigned user-defined names for identification

             * b ) **Extended numbered ACLs:**

                     - are assigned a numeric identifier, typically from the range 100 to 199 or 2000 to 2699, depending on the networking device and software used.         


 * **Re-sequencing ACLs :**

     - the process of modifying the order of access control entries (ACEs) within an Access Control List (ACL)

     -  done to change the order in which ACL rules are evaluated, which can have a significant impact on network traffic filtering and access control.

# ***NTP (Network Time Protocol)***


 * essential for maintaining accurate and synchronized time across a distributed computing environment. 

 *  NTP operates in a hierarchical structure, with a few high-stratum time servers at the top of the hierarchy.

     - These high-stratum servers obtain time from highly accurate sources, such as atomic clocks or GPS satellites.

 - It is particularly important in scenarios where precise timekeeping is critical, such as network security, financial transactions, and distributed systems.    

 * **Strata :**

     - In the NTP hierarchy, devices are classified into strata, ranging from stratum 0 (the most accurate time source) to stratum 15 (the least accurate)       

     - Stratum 1 devices are typically primary servers that synchronize their time with stratum 0 sources, and so on.

 * **NTP Servers:** 
    
     * responsible for maintaining accurate time and providing time synchronization services to clients    

 * **NTP clients :**
  
     - devices that need to synchronize their clocks with NTP servers. Clients periodically query NTP servers to adjust their time to match the server's time

# ***DHCP (Dynamic Host Configuration Protocol)***


 * protocol used in IP networks to automatically assign IP addresses and other configuration parameters to devices that connect to the network

 - DHCP simplifies the process of configuring devices on a network by automating the assignment of IP addresses, subnet masks, default gateways, DNS servers, and other network-related settings. It is commonly used in local area networks (LANs)

 - DHCP leases IP addresses to devices for a specified period. This lease duration can be customized by network administrators.

* * *


 * **D-O-R-A process in DHCP :**

         * **Discover (D):**
          
                 - a client that needs to obtain an IP address and other network settings sends out a broadcast message on the local network.

                 - This broadcast message, called a DHCP Discover packet, requests a DHCP server to provide configuration information.

         * **Offer (O):**

                 - in response to the DHCP Discover packet, DHCP servers on the network may send DHCP Offer packets.

                 - These Offer packets contain IP addresses and configuration information that the server is willing to provide to the client.

                 - The client typically receives multiple offers if multiple DHCP servers are available on the network.


         * **Request (R):**

                 - After receiving one or more DHCP Offer packets, the client evaluates the offers and selects one.

                 - The client then sends a DHCP Request packet, indicating its acceptance of the offered IP address and associated configuration options.


         * **Acknowledge (A):**

                 - In response to the DHCP Request packet, the DHCP server that the client has chosen to accept its offer sends a DHCP Acknowledge (ACK) packet.

                 - The DHCP ACK packet confirms the allocation of the requested IP address and provides the client with the selected configuration parameters.

                 - Once the client receives the ACK packet, it considers itself successfully configured and can start using the provided IP address and network settings.


         * After completing the D-O-R-A process, the client is configured with a valid IP address and other necessary network configuration parameters 

         * The client and server maintain this lease until the lease duration expires, at which point the client may initiate the process again to renew its lease or obtain a new IP address.


* * *    


 * **DHCP relay :**


         - network device or service that facilitates the exchange of DHCP messages between DHCP clients and DHCP servers in different subnets or networks

         - essential for enabling the centralized management of IP address allocation and configuration in networks with multiple subnets.

         - They help extend the reach of DHCP servers and ensure that clients in remote subnets can obtain IP addresses and network configuration information.

# ***FTP (File Transfer Protocol) & TFTP (Trivial)***


 * protocols used for transferring files between systems .

 - the both use a client-server model 

 - used to download the newer version of IOS from a server , and then reboot the device with the new IOS image 

* * *

 * **1- TFTP :**

     - named trivial because it is simple and has only basic features compared to FTP

         - only allows a client to copy a file to or from a server 


     - No authentication so servers will respond to all TFTP requests 

     * No encryption, data sent in plain text 

     * TFTP servers listen on **UDP port 69**


 * **TFTP Reliability :**

     - every TFTP data message is acknowldged 

         * if the client is transfering a file to the server , the server will send **ACK** messages  

         - if the server is transfering a file to the client , the client will send **ACK** messages    

     * timers are used , and if an expected message is not recieved in time , the waiting device will resend its previous message 

     * TFTP uses **lock-step** communication :

         * the client and server alternatively send a message and then wait for a reply and retransmission are sent as needed


 * **TFTP connections :**

     - TFTP file transfer have 3 phases :

         * 1 ) **connection :**

             - TFTP client sends a request to the server , the server responds back , initilizing the connection

         - 2 ) **Data transfer :**

             * the client and server exchange TFTP messages 

             - one sends data and the other sends acknowldgement


         * 3 ) **connection termination :**

             - after the last data message has been sent , a final acknowledgement is sent to terminate the connection      


 * **TFTP TID :**

# ***NAT (Network Address Translation)***


 * **Private IPv4 addresses :**

     - RCF (Request For Comment) :

         * documents that are used to set standards for the internet 


     * RFC specifies the following IPv4 address ranges as private :

         - 10.0.0.0/8  ( 10.0.0.0  to  10.255.255.255 )   ==> class A
         - 172.16.0.0/12  ( 172.16.0.0  to  172.31.255.255 )   ==> class B
         - 192.168.0.0/16  ( 192.168.0.0  to  192.168.255.255 )   ==> class C

     - private IP addresses cannot be used over the internet , and if you used it , two problems might  occur :

         - duplicate addresses 

         - the device cannot access the internet 

* * *
 
 * **NAT (Network Address Translation) :**
  
     - used to modify the source and/or destinationIP addresses of packets 

     - involves the translation of private (internal) IP addresses used within a local network to a single public IP address when traffic is sent to the internet

     - The translation typically occurs at a network router or firewall device that connects the private network to the public internet.


     - **Static NAT :**

         * involves statically configuring one-to-one mappings of private IP addresses to public IP addresses 

         - an inside local IP address is mapped to an inside global address :

             * **inside local :**

                 - the IP address of the inside host , from the persprctive of the local network 

                 - the IP address actyally configured on the inside host , usually a private address

             - **inside global :**

                 * the IP address of the inside host , from the perspective of outside hosts 

                 - the IP address of the inside host after NAT , usually a public address 



* * * 

 * **Static NAT configuration :**

     - 1 ) define the inside interface(s) connected to the internal network 

         * R1(config)# int g0/1
         - R1(config-if)# ip nat inside

     - 2 ) define the outside interface(s) connected to the external network  

         * R1(config)int g0/0
         - R1(config-if)# ip nat outside 

     * 3 ) configure the one-to-one IP address mappings :

         - **ip nat inside source static** inside-local-ip inside-global-ip

         * R1(config)# ip nat inside source static 192.168.0.167 100.0.0.1

         - R1(config)# ip nat inside source static 192.168.0.168 100.0.0.2     

* * * 

 * **outside local :** 

     - the IP address of the outside host , from the perspective of the local network 


 * **outside global :**

     - the IP address of the outside host , from the perspective of the outside network 

* * *


 * **Dynamic NAT :**

     - the router dynamically maps inside local addresses to inside global addresses as needed

     - An **ACL** is used to identify which traffic should be translated :

         - 1 ) if the source IP is **permitted** by the ACL , the source IP will be translted 

         - 2 ) if the source IP is **denied** by the ACL , the source IP will not be translted  (the traffic will not be dropped)

     * a NAT pool is used to define the available inside global addresses that can be used 

     - if there are not enough inside global IP addresses available , it is called **NAT pool exhaustion**

         - if a packet from another inside host arrives and needs NAT but there are no available addresses , the router will drop the packet 

         - the host will be unable to access outside networks until one of the inside global IP addresses become available 

         - Dynamic NAT entries will time out automatically if not used , or you can clear them manually 

 
 
 * **Dynamic NAT configuration :**

     - 1 ) define the inside interface(s) connected to the locla network 

          
         * R1(config)# int g0/1
         - R1(config-if)# ip nat inside

     - 2 ) define the outside interface(s) connected to the external network  

         * R1(config)int g0/0
         - R1(config-if)# ip nat outside 


     - 3 ) define the traffic that should be translated 


         - R1(config)# access-list 1 permit 192.168.0.0 0.0.0.255

     * 4 ) define the pool of inside global IP addresses 


         - R1(config)# ip nat pool POOL1 100.0.0.0 100.0.0.0 100.0.0.255 prefix-length 24

     - 5 ) config dynamic NAT by mapping the ACL to the pool 


         - R1(config)# ip nat inside list 1 pool POOl1      


* * *

 * **PAT (NAT Overload) :**

     - translate both IP address and the port number (if necessary)

     - by using a unique port number for each communication flow , a single public IP address can be used by many different internal hosts 

     - the router will keep track of which inside local address is using which inside global address and port 


 * **PAT configuration (pool) :**

     - 1 ) define the inside interface(s) connected to the locla network 

          
         * R1(config)# int g0/1
         - R1(config-if)# ip nat inside

     - 2 ) define the outside interface(s) connected to the external network  

         * R1(config)int g0/0
         - R1(config-if)# ip nat outside 


     - 3 ) define the traffic that should be translated 


         - R1(config)# access-list 1 permit 192.168.0.0 0.0.0.255

     * 4 ) define the pool of inside global IP addresses 


         - R1(config)# ip nat pool POOL1 100.0.0.0 100.0.0.0 100.0.0.3 prefix-length 24

     - 5 ) config PAT by mapping the ACL to the pool and using the **overload** keyword at the end


         - R1(config)# ip nat inside list 1 pool POOl1 overload     


* **PAT configuration (interface) :**

     - 1 ) define the inside interface(s) connected to the locla network 

          
         * R1(config)# int g0/1
         - R1(config-if)# ip nat inside

     - 2 ) define the outside interface(s) connected to the external network  

         * R1(config)int g0/0
         - R1(config-if)# ip nat outside 


     - 3 ) define the traffic that should be translated 


         - R1(config)# access-list 1 permit 192.168.0.0 0.0.0.255

     
     - 4 ) config PAT by mapping the ACL to the interface and enabling overload 


         - R1(config)# ip nat inside source list 1 interface g0/0 overload     




# ***QoS (Quality of Service)***


 * **IP phones :**

     - traditional phones operate over the Public Switced Telephone Network (PSTN) , aslo called POTS (Plain Old Telephone Service)

     - IP phones use VoIP (Voice over IP) technologies to enble phone calls over IP network , such as the internet 

     - IP phones are connecteed to a switch just like any other end host 

     - IP phones have an internal 3-port switch :

         - 1 port is the **uplink** to the external switch 

         - 1 port is the **downlink** to the PC

         - 1 port connects internally tothe phone itself

     - this allows the PC and the IP phone to share a single switch port . traffic from the Pc passes through the IP phone to the switch


     - it is recommended to separate voice traffic (from the IP phone) and data traffic (from the PC) by placing them in separate VLANs 

         - this can be accomplished using voice VLAN

         - traffic from the PC will be untagged , but traffic from the phone will be tagge with a VLAN ID


* * *

 * **PoE (Power over Ethernet) :**

     - allows Power Sourcing Equipment (PSE) such as a switch to provide power to Powered Devices (PD) such as IP phones over an Ethernet cable 

     - the PSE recieves AC power from the outlet , converts it to DC power , and supplies that DC power to the PDs


     - PoE has a process to determine if a connected device needs power  , and how much power it needs :

         - when a device is connected to a PoE-enabled port , the PSE (switch) sends low power signals , monitors the response , and determines how much power the PD needs 

         - if the device needs power , the PSE supplies the power to allow the PD to boot 

         - the PSE continues to monitor the PD and supply the required amount of power 

* * *

* Voice traffic uses the PSTN 

* Data traffic uses the IP network (enterprise WAN, internet, etc)


* * *

 - **QoS :**


     * is a set of tools used by network devices to apply different treatment to different packets 

     - QoS is used to managed the following characteristics of network traffic :

         - 1 ) **Bandwidth :**

             - the overall capacity of the link , measured in bits per seconds

             - QoS tools allows you to reserve a certain amount of a link’s bandwidth for specific kinds of traffic 

         * 2 ) **Delay :**

             - the amount of time it takes traffic to go from source to destination ==> **one-way delay**   

                 - for acceptable interactive audio quality **(one-way delay of 150 ms or less)** is recommended 

             -  the amount of time it takes traffic to go from source to destination and return ==> **two-way delay** 


         - 3 ) **Jitter :**

             - the variation in one-way delay between packets sent by the same application 

             - IP phones have a **jitter buffer** to provide a fixed delay to audio packets 

             -  for acceptable interactive audio quality **(jitter 30 of ms or less)** is recommended 



         * 4 ) **Loss :**

             - the % of packets sent that do not reach their destination 

             - can be caused by faulty cables 

             - can also be caused when a device’s packet queues get full and the device starts discarding packets 

             -  for acceptable interactive audio quality **(loss of 1% or less)** is recommended 


* * *


 * **QoS - queuing**

     - if a network device recieves messages faster than it can forward them out the appropriate interface , the message are placed in a **queue**

     - by default , queued messages will be forwarded in a **First In First Out (FIFO)** manner 

         - messages will be sent in the order they are recieved 

     - if the queue is full , new packets will be dropped **(tail drop)** which is harmful because it can lead to :

         * **TCP global syncronization :**

             - Review of the **TCP sliding window :**

                 - hosts using TCP use the **sliding window** increase/decrease the rate at which they send traffic as needed 

                 - when a packet is dropped it will be re-transmitted 

                 - when a drop occurs , the sender will reduce the rate it send traffic

                 - it will then gradually increase the rate again 

             - when the queue fills up and tail occurs , all TCP hosts sending traffic will slow down the rate at which they send traffic

             - they will all then increase the rate a which they send traffic , which rapidly leads to more congestion , dropped packets , and the process repeats again 


         * to prevent TCP global syncronization we use **RED (Random Early Detection) :** 

             - when the amount of traffic in the queue reaches a certain threshold , the device will start randomly dropping packets from select TCP flows 

             - those TCP flows that dropped packets will reduce the rate at which traffic is sent , but you will avoid global TCP syncronization 

             - an improved version , **WRED (Weighted Random Early Detection)** , allows you to control which packets are dropped depending on the traffic class    


* * *

* **Classificaion :**

 - organizes network traffic into trffic classes 

 - fundamentla to QoS to give priority to certain types of traffic and you have to identify which types of traffic to give priority to 

 - **methods of classifying traffic :**

     - 1 ) **PCP (Priority Code Point) or CoS (Class of Service) :**

         - its use is defined by IEEE 802.1p

         - there are 3 bits which gives 8 possible PCP values :

             - a ) **0 :**
                 - traffic type ==> best effort (means there is guarantee that data is delivered or that it meets any QoS standard . this is regular traffic , not high priority)

             - b ) **1 :**
                 - traffic type ==> background 

             - c ) **2 :**
                 - traffic type ==> excellent effort 

             - d ) **3 :**
                 - traffic type ==> critical applications 

             - e ) **4 :**
                 - traffic type ==> video 

             - f ) **5 :**
                 - traffic type ==> Voice 

             - g ) **6 :**
                 - traffic type ==> interconnection control 

             - h ) **7 :**
                 - traffic type ==> network control

         * it can only be used over the following connections :

             - 1- trunk links 
             - 2- access links with voice VLAN
